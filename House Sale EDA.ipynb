## Analysis Descriptive

Analisis deskriptif untuk harga
1. mean , median, modus
2. Q1,Q2, dan interquartile 
3. range, standar deviasi, dan variance

```python
print ('Nilai rata rata harga :',round(df['price'].mean(),2))
print ('Nilai median harga :',df['price'].median())
print ('Nilai modus harga :',df['price'].mode())
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
interquartile = Q3-Q1
print ('Nilai Q1 harga :',Q1)
print ('Nilai Q1 harga :',Q3)
print ('Nilai Interquartile harga :',Q3-Q1)
```

```python
price_max = df['price'].max()
price_min = df['price'].min()
price_range = price_max-price_min

print ('Nilai Harga Terendah',price_min)
print ('Nilai Harga Tertinggi',price_max)
print ('Nilai Range Harga :', price_range)
print ('Nilai Standar Deviasi Harga :', round(df['price'].std(),2))
print ('Nilai Variance Harga ;', round(df['price'].var(),2))
```

## Kausalitas antara harga dengan Grade

- Nama penelitian : Hubungan kausalitas antara harga dengan Grade Rumah
- matrix yang digunakan : harga dan grade
- uji normalitas : menggunakan shapiro wilk dan kolmogorof smirnov untuk membandingkan keduanya
- uji kausalitas : menggunakan biserial karena 1 data ordinal (grade) dan 1 data ratio

## Uji normalitas 

1. Shapiro Wilk

```python
from scipy.stats import shapiro #mengambil uji shapiro wilk pada scipy.stats

H0 = 'Data terdistribusi Normal atau terima H0'
H1 = 'Data tidak terdistribusi Normal atau terima H1'

stat, p_value_sw = shapiro(df['price']) #uji normalitas shapiro wilk

print (stat, p_value_sw)

if p_value_sw < 0.05:
    print (H0)
else :
    print (H1)
```

2. Uji Kolmogorov Smirnov

```python
from scipy.stats import kstest #mengambil uji kolmogorov smirnov pada scipy.stats

H0 = 'Data terdistribusi Normal atau terima H0'
H1 = 'Data tidak terdistribusi Normal atau terima H1'

stat, p_value_ks = kstest(df['price'],'norm') #uji normalitas kolmogorov smirnov

print (stat, p_value_ks)

if p_value_sw < 0.05:
    print (H0)
else :
    print (H1)
```
